```python
import numpy as np
import pandas as pd
```

```python
LOG_PATH = "/tmp/log/"

latency = np.loadtxt(LOG_PATH + "latency_log.txt")
throughput = np.loadtxt(LOG_PATH + "throughput_log.txt")

print(f"average latency per block of sensors: {np.average(latency) * 17} seconds")
print(f"average throughput per block of sensors: {np.average(throughput) * 17} KB")
```

```python
print("Latency line plot in seconds")
pd.Series(latency).plot.line(figsize=(20,13));
```

```python
print("Throughput line plot in KB")
pd.Series(throughput).plot.line(figsize=(20,13));
```
